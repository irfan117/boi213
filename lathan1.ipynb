```python
import matplotlib.pyplot as plt
import numpy as np

propinsi = ['Jawa Barat', 'Jawa Tengah', 'Jawa Timur', 'DKI Jakarta', 'DI Yogyakarta', 'Banten', 'Bali', 'Nusa Tenggara Timur',  'Nusa Tenggara Barat']
jumlah_penduduk = [49565200, 34738200, 39955900, 10576400, 3919200, 12895300, 4414400, 5513400, 5225900]

x_coords = np.arange(len(propinsi))
plt.bar(
    x_coords,
    jumlah_penduduk,
    tick_label= propinsi
)

plt.xticks(rotation=90)
plt.ylabel('Jumlah Penduduk')
plt.title('Data Penduduk Jawa, Bali. NTT, dan NTB Tahun 2020')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

propinsi = ['Jawa Barat', 'Jawa Tengah', 'Jawa Timur', 'DKI Jakarta', 'DI Yogyakarta', 'Banten', 'Bali', 'Nusa Tenggara Timur',  'Nusa Tenggara Barat']
jumlah_penduduk = [49565200, 34738200, 39955900, 10576400, 3919200, 12895300, 4414400, 5513400, 5225900]

data_penduduk = pd.DataFrame({'Propinsi': propinsi, 'Jumlah Penduduk': jumlah_penduduk})
data_penduduk.sort_values(by='Jumlah Penduduk', inplace=True)

x_coords = np.arange(len(propinsi))
plt.bar(
    x_coords,
    data_penduduk['Jumlah Penduduk'],
    tick_label= data_penduduk['Propinsi']
)

plt.xticks(rotation=90)
plt.ylabel('Jumlah Penduduk')
plt.title('Data Penduduk jawir, Bali. NTT, dan NTB Tahun 2020')

plt.show()
```

```python
warna = ['#0000FF', '#0000FF', '#0000FF', '#0000FF', '#0000FF', '#0000FF', '#0000FF', '#FF0000', '#0000FF']

x_coords = np.arange(len(propinsi))
plt.bar(
    x_coords,
    data_penduduk['Jumlah Penduduk'],
    tick_label= data_penduduk['Propinsi'],
    color = warna
)

plt.xticks(rotation=90)
plt.ylabel('Jumlah Penduduk')
plt.title('Data Penduduk Jawa, Bali. NTT, dan NTB Tahun 2020')

plt.show()
```

#Diagram Garis

<b>Data Pencatatan Jumlah Pengunjung Restoran X di tanggal 22 Agustus 2021</b>

Asumsi: Data dicatat mulai dari jam 6 pagi sampai dengan jam 21. Pencatatan data dicatat sekali dalam 1 jam.

100, 102, 98, 70, 300, 305, 546, 690, 120, 145, 43, 57, 600, 987, 800, 234

```python
jam_catat = [x for x in range(6,22)]
jumlah_pengunjung = [100, 102, 98, 70, 300, 305, 546, 690, 120, 145, 43, 57, 600, 987, 800, 234]

plt.plot(
    jam_catat,
    jumlah_pengunjung
)

plt.title('Data Jumlah Pengunjung Restoran Kemiri di tanggal 22 Agustus 2025')
plt.ylabel('Jumlah Pengunjung')
plt.xlabel('Jam Pencatatan')
plt.show()
```

menambahkan marker

```python
plt.plot(
    jam_catat,
    jumlah_pengunjung,
    marker = 'x'
)

plt.title('Data Jumlah Pengunjung Restoran Kemiri di tanggal 22 Agustus 2025')
plt.ylabel('Jumlah Pengunjung')
plt.xlabel('Jam Pencatatan')
plt.show()
```

Contoh perbandingan dalam diagram garis

di bawah contoh kasus perbandingan jumlah pengunjung data aktua dan prediksi

```python
jumlah_pengunjung_prediksi = [90, 100, 98, 90, 296, 300, 500, 500, 127, 154, 50, 40, 700, 966, 845, 100]

plt.plot(
    jam_catat,
    jumlah_pengunjung,
    marker = 'x',
    label = 'Aktual'
)

plt.plot(
    jam_catat,
    jumlah_pengunjung_prediksi,
    linestyle = '--',
    marker = 'x',
    label = 'Prediksi'
)

plt.legend()
plt.title('Data Jumlah Pengunjung (Aktual dan Prediksi) Restoran Kemiri di tanggal 22 Agustus 2025')
plt.ylabel('Jumlah Pengunjung')
plt.xlabel('Jam Pencatatan')
plt.show()
```
